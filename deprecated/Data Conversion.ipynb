```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd
import numpy as np
```

# Transform Data into CSV files

```python
# read original text files
names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
         'occupation', 'relationship', 'race', 'sex',  'capital-gain', 'capital-loss', 
         'hours-per-week', 'native-country', 'target']

train = pd.read_csv('./data/adult.data.txt', sep=', ', header=None, names=names)
test = pd.read_csv('./data/adult.test.txt', sep=', ', header=None, names=names)
```

```python
# replace '?' with np.nan
train = train.replace('?', value=np.nan)
test = test.replace('?', value=np.nan)
```

```python
# process test target: '<=50K.' ==> '<=50K', '>50K.' ==> '>50K'
test['target'] = test['target'].apply(lambda x: x[:-1])
```
